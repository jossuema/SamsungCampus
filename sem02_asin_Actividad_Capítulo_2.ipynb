# Actividad: Creación de una Base de Datos en Python

## Descripción

A continuación debes crear una base de datos simple utilizando listas en Python. Una base de datos es una colección de datos organizados que se utilizan para almacenar y recuperar información. En este caso, simularemos una base de datos de estudiantes con información básica.

## Instrucciones

1. Crea un programa en Python que permita a los usuarios realizar las siguientes acciones:

   a. Agregar un nuevo estudiante a la base de datos. Cada estudiante debe tener al menos un nombre, un número de identificación y una edad. Puedes utilizar una lista para almacenar estos datos.

   b. Mostrar la lista de estudiantes en la base de datos.

   c. Buscar un estudiante por su número de identificación y mostrar su información.

   d. Actualizar la información de un estudiante existente en la base de datos.

   e. Eliminar un estudiante de la base de datos.

2. Utiliza listas de Python para almacenar la información de los estudiantes. Por ejemplo, puedes crear una lista de diccionarios, donde cada diccionario representa un estudiante con sus datos.

3. Ofrece un menú interactivo que permita a los usuarios elegir las acciones que desean realizar en la base de datos.

4. Implementa la lógica necesaria para cada acción, como agregar, buscar, actualizar o eliminar estudiantes.

## Ejemplo de Base de Datos

A continuación, se muestra un ejemplo de cómo podría verse una base de datos de estudiantes utilizando listas y diccionarios:

```python
base_de_datos = [
    {"nombre": "Juan", "id": 1, "edad": 20},
    {"nombre": "María", "id": 2, "edad": 22},
    {"nombre": "Carlos", "id": 3, "edad": 21}
]

Desarrolle el código en la siguiente lista.

````python
# Base de datos simulada
base_de_datos = [
    {"nombre": "Juan", "id": 1, "edad": 20},
    {"nombre": "María", "id": 2, "edad": 22},
    {"nombre": "Carlos", "id": 3, "edad": 21}
]

# Función para agregar un estudiante
def agregar_estudiante():
    nombre = input("Ingrese el nombre del estudiante: ")
    id_estudiante = int(input("Ingrese el número de identificación del estudiante: "))
    edad = int(input("Ingrese la edad del estudiante: "))
    estudiante = {"nombre": nombre, "id": id_estudiante, "edad": edad}
    base_de_datos.append(estudiante)
    print("Estudiante agregado correctamente.\n")

# Función para mostrar todos los estudiantes
def mostrar_estudiantes():
    if len(base_de_datos) == 0:
        print("No hay estudiantes en la base de datos.\n")
    else:
        print("Lista de estudiantes:")
        for estudiante in base_de_datos:
            print(f"ID: {estudiante['id']}, Nombre: {estudiante['nombre']}, Edad: {estudiante['edad']}")
        print()

# Función para buscar un estudiante por su ID
def buscar_estudiante():
    id_estudiante = int(input("Ingrese el número de identificación del estudiante que desea buscar: "))
    encontrado = False
    for estudiante in base_de_datos:
        if estudiante["id"] == id_estudiante:
            print(f"Estudiante encontrado: ID: {estudiante['id']}, Nombre: {estudiante['nombre']}, Edad: {estudiante['edad']}\n")
            encontrado = True
            break
    if not encontrado:
        print("Estudiante no encontrado.\n")

# Función para actualizar la información de un estudiante
def actualizar_estudiante():
    id_estudiante = int(input("Ingrese el número de identificación del estudiante que desea actualizar: "))
    encontrado = False
    for estudiante in base_de_datos:
        if estudiante["id"] == id_estudiante:
            nuevo_nombre = input("Ingrese el nuevo nombre: ")
            nueva_edad = int(input("Ingrese la nueva edad: "))
            estudiante["nombre"] = nuevo_nombre
            estudiante["edad"] = nueva_edad
            print("Información actualizada correctamente.\n")
            encontrado = True
            break
    if not encontrado:
        print("Estudiante no encontrado.\n")

# Función para eliminar un estudiante
def eliminar_estudiante():
    id_estudiante = int(input("Ingrese el número de identificación del estudiante que desea eliminar: "))
    encontrado = False
    for estudiante in base_de_datos:
        if estudiante["id"] == id_estudiante:
            base_de_datos.remove(estudiante)
            print("Estudiante eliminado correctamente.\n")
            encontrado = True
            break
    if not encontrado:
        print("Estudiante no encontrado.\n")

# Función para mostrar el menú
def mostrar_menu():
    print("Menú de opciones:")
    print("1. Agregar un nuevo estudiante")
    print("2. Mostrar lista de estudiantes")
    print("3. Buscar un estudiante")
    print("4. Actualizar información de un estudiante")
    print("5. Eliminar un estudiante")
    print("6. Salir")

# Función principal que ejecuta el programa
def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción (1-6): ")
        
        if opcion == "1":
            agregar_estudiante()
        elif opcion == "2":
            mostrar_estudiantes()
        elif opcion == "3":
            buscar_estudiante()
        elif opcion == "4":
            actualizar_estudiante()
        elif opcion == "5":
            eliminar_estudiante()
        elif opcion == "6":
            print("Gracias por usar el sistema. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.\n")

# Ejecutar el programa
if __name__ == "__main__":
    main()
````


# Actividad: Diferentes métodos usados en listas.

De la unidad 10, sección 6.3 revise los métodos disponibles para usar sobre las listas.

Aplique los métodos disponibles sobre la lista mostrada en la siguiente celda. 

Nota: usando el método **sort()** ordene de mayor a menor y viceversa (Si es necesario revise documentación adicional relacionada sobre este método)


````python
list_notas = [15, 21, 24, 40, 13, 14, 18, 17, 14.5]

# Agregar un nuevo elemento
list_notas.append(19)
print("Después de append:", list_notas)

# Extender la lista
list_notas.extend([30, 22, 10])
print("Después de extend:", list_notas)

# Insertar un nuevo elemento
list_notas.insert(3, 50)
print("Después de insert:", list_notas)

# Eliminar el primer elemento con valor específico
list_notas.remove(40)
print("Después de remove:", list_notas)

# Eliminar y devolver el último elemento
ultimo_elemento = list_notas.pop()
print("Elemento eliminado con pop:", ultimo_elemento)
print("Después de pop:", list_notas)

# Obtener el índice de un valor
indice = list_notas.index(50)
print("Índice de 50:", indice)

# Contar cuántas veces aparece un valor
count_14 = list_notas.count(14)
print("Cantidad de veces que aparece 14:", count_14)

# Ordenar de menor a mayor
list_notas.sort()
print("Después de ordenar de menor a mayor:", list_notas)

# Ordenar de mayor a menor
list_notas.sort(reverse=True)
print("Después de ordenar de mayor a menor:", list_notas)
````

# Actividad: Crear diccionarios y guardarlos como archivo JSON.

De la unidad 11 sección 4.6 revise el ejemplo mostrado.

Genera un diccionario, sobre la información que deseas guardar como base de datos, desarrollo el código necesario para generar el archivo JSON, y también el código para cargar la información del archivo JSON.


````python
import json

# Crear un diccionario con la información de los estudiantes
estudiantes = {
    "estudiantes": [
        {"id": 1, "nombre": "Juan", "edad": 20, "carrera": "Ingeniería"},
        {"id": 2, "nombre": "María", "edad": 22, "carrera": "Medicina"},
        {"id": 3, "nombre": "Carlos", "edad": 21, "carrera": "Derecho"},
        {"id": 4, "nombre": "Ana", "edad": 23, "carrera": "Arquitectura"}
    ]
}

# Guardar el diccionario como un archivo JSON
with open('estudiantes.json', 'w') as file:
    json.dump(estudiantes, file, indent=4)

print("Diccionario guardado en el archivo estudiantes.json")
````

````python
# Leer el archivo JSON y cargar los datos en un diccionario
with open('estudiantes.json', 'r') as file:
    datos_cargados = json.load(file)

# Mostrar los datos cargados
print("Datos cargados desde el archivo JSON:")
print(datos_cargados)
````

# Actividad: Revisión y manejo de listas bidimensionales.

## Instrucciones

1. Describa brevemente que entiende por lista bidimensional.
2. Crear una lista bidimensional.
3. Generar un doble loop para acceder de la lista bidimensional
4. Sumar todo los elementos de esta lista bidimensional.

````python
# Lista bidimensional con calificaciones
calificaciones = [
    [85, 90, 88, 92],  # Estudiante 1
    [78, 82, 80, 84],  # Estudiante 2
    [92, 95, 94, 96]   # Estudiante 3
]

# Doble loop para acceder a los elementos de la lista bidimensional
print("Calificaciones de los estudiantes:")
for fila in calificaciones:
    for calificacion in fila:
        print(calificacion, end=" ")
    print()  # Salto de línea al final de cada fila

# Sumar todos los elementos de la lista bidimensional
suma_total = 0
for fila in calificaciones:
    for calificacion in fila:
        suma_total += calificacion

print("La suma total de las calificaciones es:", suma_total)
````

# Actividad: Revision de los métodos usados en el manejo de diccionarios.

## Instrucciones

1. Crear un diccionario.
2. Testear los siguientes métodos para el majeo de diccionarios get(), keys(), values(), items(), pop(), clear()
3. Agregar un breve comentario sobre su funcionamiento.

````python
# Crear un diccionario con información de empleados
empleados = {
    "Juan": {"edad": 30, "departamento": "Ventas"},
    "María": {"edad": 28, "departamento": "Marketing"},
    "Carlos": {"edad": 35, "departamento": "Finanzas"},
    "Ana": {"edad": 40, "departamento": "RRHH"}
}

# get()
print(empleados.get("Juan"))
print(empleados.get("Pedro", "Empleado no encontrado"))

# keys()
print(empleados.keys())

# values()
print(empleados.values())

# items()
for clave, valor in empleados.items():
    print(clave, ":", valor)

# pop()
print(empleados.pop("María"))
print(empleados.pop("Pedro", "Clave no encontrada"))

# clear()
empleados.clear()
print(empleados)
````
